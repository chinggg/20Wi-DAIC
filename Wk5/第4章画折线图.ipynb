```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
data = DataFrame([41.3,48.9,54,59.5,64.4,68.9,74.4],
                 index=['2010','2011','2012','2013','2014','2015','2016'],
                 columns=['GDP:Trillion'])
data
```

```python
data.plot(marker='*')
plt.show()
```

```python
#画精细的折线图
data.plot(marker='*',color='r')
plt.title('2010-2016 GDP')
plt.xlabel('year')
plt.ylabel('GDP', fontsize=12)
plt.ylim(35,80)
plt.grid()
plt.savefig('GDP.jpg', dpi=200)
plt.show()
```
