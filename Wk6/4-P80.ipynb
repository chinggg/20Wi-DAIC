```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
data = pd.read_csv('data/High-speed rail.csv',index_col=0,usecols=[0,1,2,3])
data
```

```python
data['Operation'].plot(kind='bar', xlabel='Country', title='Operation Mileage')
plt.annotate('Longest!',xy=(0,20000),xytext=(1,20000),arrowprops=dict(arrowstyle='->'))
```

```python
data.plot(kind='barh', stacked=True, title="Global trends of high-speed rail", 
          xlabel='Mileage(km)', ylabel='Country')
```

```python
data['Operation'].plot(kind='pie', title='Operation Mileage', ylabel='', 
               shadow=True, explode=[0.1,0,0,0,0,0], startangle=60, autopct='%1.1f%%')
```
