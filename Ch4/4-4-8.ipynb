```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'data/GDP.csv', index_col='Year')
# use 'linewidth' instead of 'LineWidth'
data.plot(title='GDP & Income', linewidth=2, marker='o',
         linestyle='dashed', grid=True, use_index=True)
```

```python
# 绘制半对数折线图
data.plot(logy=True, linewidth=2, marker='o', linestyle='dashed', color='g')
```
