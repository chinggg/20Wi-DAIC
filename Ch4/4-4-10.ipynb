```python
import pandas as pd
```

```python
data = pd.read_csv(r'data/advertising.csv')
piedata = data[['TV','Weibo','WeChat']]
datasum = piedata.sum()
```

```python
# 绘制饼图
datasum.plot(kind='pie', figsize=(6,6), title='Advertising Expenditure',
fontsize=14, explode=[0,0.2,0],shadow=True, startangle=60, autopct='%1.1f%%')
```
