```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
data = DataFrame([1.47,1.62,1.78,1.94,2.38,2.60],
                 index=['2012','2013','2014','2015','2016','2017'],
                 columns=['Income'])
data
```

```python
data.plot(marker='s', linestyle='--', color='k')
plt.title('Income chart')
plt.xlabel('year')
plt.ylabel('Income(RMB Ten Thousand)')
plt.yticks(np.arange(0,3.1,0.5))
plt.annotate('Largest!',xy=(5,2.6),xytext=(3,2.5),arrowprops=dict(arrowstyle='->',color='r'),color='r')
plt.savefig('2012-2017Income.jpg', dpi=400, bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
ax1 = fig.add_subplot(2,2,1)
data.plot(title='linechart', xlabel='Year', ylabel='Income',
          yticks=np.arange(1.5,2.6,0.25), ax=ax1)
ax2 = fig.add_subplot(2,2,2)
data.plot(kind='box', title='box-whisker plot',
        ylabel='Income', xticks=[], yticks=np.arange(1.5,2.6,0.25), ax=ax2)
plt.xlabel('2012â€”2017')  #bug
ax3 = fig.add_subplot(2,1,2)
data.plot(kind='bar', title='bar chart', xlabel='Year', ylabel='Income', ax=ax3)
plt.subplots_adjust(wspace=0.4, hspace=0.4)
```
