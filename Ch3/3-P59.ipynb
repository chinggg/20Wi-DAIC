```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
data = pd.read_excel('data/DataScience.xls')
data
```

```python
# 查询实验教学计划的基本内容及总数
data.shape
```

```python
# 查询实验教学计划表中是否含有NaN数据。
print(data.isnull().any())
# 将含有NaN数据的行导出为数据文件pre.csv
t = data.loc[data.isnull().any(axis=1)]
t.to_csv('pre.csv')
# 判断采用何种数据清洗模式：填充、删除或手工填充；
# 由于课表排布非常规则，仅需删除全为NaN的行，其余缺失项用前一行填充即可
data.dropna(how='all', inplace=True)
data.fillna(method='ffill', inplace=True)
data.isnull().any()
```

```python
# 查询课程名称、实验项目名称、实验类型和二级实验室四列数据内容
data[['课程名称','实验项目名称','实验类型','二级实验室名称']]
```

```python
# 统计每一门课程的实验课时数
# 按照‘课程名称’分组, 计算 ‘实验课时数’ 的和
name_grouped = data.groupby('课程名称')
name_sum = name_grouped.aggregate({'实验课时数':np.sum})
name_sum
```

```python
# 统计每周开设的各门实验课的时数
# 按照 '周次', '课程名称' 两列数据分组 
wknm_grouped = data.groupby(['周次','课程名称'])
wknm_sum = wknm_grouped.aggregate({'实验课时数':np.sum})
wknm_sum
```

```python
# 统计每门课程的实验类型分布（crosstab）
pd.crosstab(data['课程名称'],data['实验类型'])
```

```python
# 统计每个班级的实验课课表
# （1）将数据按照 ['班级','周次','星期','节次']  这4个列排序（升序）
#（2）再选出这些列['班级','周次','星期','节次','课程名称','实验项目名称']
data.sort_values(by=['班级','周次','星期','节次'], inplace=True)
data[['班级','周次','星期','节次','课程名称','实验项目名称']]
```

```python
# 分析各二级实验室承担的实验课时量
lab_grouped = data.groupby('二级实验室名称')
lab_sum = lab_grouped.aggregate({'实验课时数':np.sum})
lab_sum
```
