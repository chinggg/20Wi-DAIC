```python
# 数据集bankdebt.csv
# 1. 将数据集划分为训练集和测试集，查看决策树分类器的性能。
# 2. 将例5-3学习的分类器保存到文件中，然后重新加载预测给出的新数据
import pandas as pd
data = pd.read_csv('../data/bankdebt.csv', index_col=0, header=None)
data
```

```python
dic = {'Yes':1, 'No':0, 'Single':1, 'Married':2, 'Divorced':3}
data.replace(dic, inplace=True)
data
```

```python
X = data.loc[:, 1:3].values.astype(float)
Y = data.loc[:, 4].values.astype(int)
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.35, random_state=1)
from sklearn import tree
clfTr = tree.DecisionTreeClassifier()
clfTr = clfTr.fit(X_train, Y_train)
print('分类器在测试集上的的分数: ', clfTr.score(X_test, Y_test), '\n')
predicted_Y = clfTr.predict(X_train)
from sklearn import metrics
print(metrics.classification_report(Y_train, predicted_Y))
print('混淆矩阵:')
print(metrics.confusion_matrix(Y_train, predicted_Y))
```

```python
# 将分类器保存到文件
import joblib

joblib.dump(clfTr,'myclf')
```
